```python
X = df_clean[['Developer','Age','Gender','Area','Immigrant','English']]
X.head()
```

```python
y = df_clean[['HighIncome']]
y.head()
```

```python
education_level_order=[[
    'No high school (secondary school)',
    'Some high school',
    'High school diploma or equivalent (GED)',
    'Some college credit, no degree',
    'Associate‚Äôs degree',
    'Bachelor‚Äôs degree',
    "Master's degree (non-professional)",
    'Professional degree (MBA, MD, JD, etc.)',
    'Ph.D.',
    'Trade, technical, or vocational training'
]]
from sklearn.preprocessing import OrdinalEncoder
encoder = OrdinalEncoder(categories = education_level_order)
X['EducationLevel'] = encoder.fit_transform(df_clean[['Education']])
display(X)
```

```python
X = pd.get_dummies(X) 
X.head()
```

```python
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3) 
print(X_train[0:5])
```

```python
from sklearn import tree 
from sklearn import metrics 
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix,ConfusionMatrixDisplay, roc_curve, auc
#model1= tree.DecisionTreeClassifier()
model1 = tree.DecisionTreeClassifier(criterion='gini', splitter='best', max_depth=5, min_samples_split=20, min_samples_leaf=20, min_weight_fraction_leaf=0.0, max_features=None, random_state=None, max_leaf_nodes=None, min_impurity_decrease=0.0, class_weight=None, ccp_alpha=0.0)  #build a decision tree
```

```python
model1 = model1.fit(X_train, y_train) #fit decision tree
```

```python
y_pred = model1.predict(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

param_grid = {'criterion':["gini","entropy"],
'max_depth': [2, 3, 4, 5],
'min_samples_leaf': [10, 20, 30],
'min_samples_split': [20, 30, 40]}

clf = DecisionTreeClassifier()

grid_search = GridSearchCV(clf, param_grid, cv=5)

grid_search.fit(X_train, y_train)

print("Best hyperparameters: ", grid_search.best_params_)
```

```python
Accuracy = metrics.accuracy_score(y_test, y_pred) 
print ("Accuracy: ", Accuracy) 
```

```python
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
from sklearn.tree import export_graphviz
import graphviz

export_graphviz(model1, out_file='tree.dot', 
                feature_names=X.columns,  
                class_names=["Low", "High"],  
                filled=True, rounded=True,  
                special_characters=True)

# Read the exported .dot file and render it as a Graphviz object
with open("tree.dot") as f:
    dot_data = f.read()
graph = graphviz.Source(dot_data)

# Display the decision tree
display(graph)
```
